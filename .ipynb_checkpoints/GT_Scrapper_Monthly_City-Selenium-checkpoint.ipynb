```python
import pandas as pd
import numpy as np
import requests
import time
import random
from datetime import datetime
import csv
from webdriver_manager.chrome import ChromeDriverManager
import os
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.webdriver.common.action_chains import ActionChains
```

# Functions

```python
def get_web_driver():
    global driver
    service = Service()
    options = webdriver.ChromeOptions()
    #options.add_argument("--incognito")
    driver = webdriver.Chrome(service = service, options = options)
    driver.get("https://trends.google.com/trends/")
    time.sleep(2)
```

```python
def click_and_scroll():

    element = driver.find_element("xpath","/html/body/div[2]/div[2]/div/md-content/div/div/div[1]/trends-widget/ng-include/widget/div/div/div/div")

    # Click on the element to set focus
    ActionChains(driver).move_to_element(element).click().perform()

    # Scroll down the page
    driver.execute_script("window.scrollBy(0, 500)")
```

```python
def column_name_to_mvt_and_remove_zero_values(df, geo_name):
    df.rename(columns={df.columns[0]: f"MVT_GT_{geo_name}" }, inplace = True)
    df = df[df[f"MVT_GT_{geo_name}"] != 0]
    return df
```

```python
def click_time_range_button(): 
    click_time_range = True
    while click_time_range == True:
        try:   
            time.sleep(1)
            print("here1")
            driver.find_element("xpath","/html/body/div[2]/div[2]/div/header/div/div[3]/ng-transclude/div[2]/div/div/custom-date-picker").click()
            click_time_range = False
        except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:    
            time.sleep(0.5)
            try:  
                time.sleep(1)
                print("here2")
                driver.find_element("xpath","/html/body/div[2]/div[2]/div/header/div/div[3]/ng-transclude/div[2]/div/div/custom-date-picker/ng-include").click()
                click_time_range = False
            except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:      
                time.sleep(0.5)
                driver.refresh()
                print("here3")
                time.sleep(2)
                continue
```

```python
 def click_custom_range_button():     
    click_custom_range = True
    while click_custom_range == True:
        try: 
            time.sleep(1)
            print("here4")               
            driver.find_element("xpath","/html/body/div[8]/md-select-menu/md-content/md-option[10]").click()
            click_custom_range = False
        except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:
            time.sleep(0.5)
            driver.refresh()
            time.sleep(2)
            try:
                time.sleep(1)
                print("here5")
                driver.find_element("xpath","/html/body/div[7]/md-select-menu/md-content/md-option[10]").click()
                click_custom_range = False
            except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:
                try:
                    time.sleep(0.5)
                    driver.refresh()
                    time.sleep(2)
                    print("here6")               
                    driver.find_element("xpath","/html/body/div[9]/md-select-menu/md-content/md-option[10]").click()
                    click_custom_range = False
                except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:
                    try:
                        time.sleep(0.5)
                        driver.refresh()
                        time.sleep(2)
                        print("here7")
                        click_time_range_button()
                    except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:
                        time.sleep(0.5)
                        driver.refresh()
                        time.sleep(2)
                        print("here8")
                        continue
```

```python
def get_time_series_metro_GT_csv(search_keywords, time_start, time_end, geo_code):
    from selenium.webdriver.common.keys import Keys
    #get the first screen on GT and delete the defult search words
    get_web_driver()
    
    for i in range(0, 50):
        driver.find_element("xpath","/html/body/c-wiz/div/div[2]/div[4]/div[1]/c-wiz[1]/div/div[1]/div[3]/div/div/div[2]/div/label/span[2]/input").send_keys(Keys.BACKSPACE)
        time.sleep(0.05)
        
    #Send search keys
    driver.find_element("xpath","/html/body/c-wiz/div/div[2]/div[4]/div[1]/c-wiz[1]/div/div[1]/div[3]/div/div/div[2]/div/label/span[2]/input").send_keys(search_keywords)
    time.sleep(1)
    
    #Click the "Explore" button
    driver.find_element("xpath","/html/body/c-wiz/div/div[2]/div[4]/div[1]/c-wiz[1]/div/div[1]/div[3]/div/div[2]/div[3]/div/button/span[6]").click()
    time.sleep(1)
    driver.maximize_window()
    time.sleep(2)
    
    #Select Custom Date Range
    #I write two special function to deal with this button as the full xpath of its keep jumping from one to another
    click_time_range_button()
    click_custom_range_button()
    time.sleep(1)
    
    #Clear and Send "from" date
    for i in range(0, 11):
        try:
            driver.find_element("xpath", "/html/body/div[2]/div[4]/md-dialog/md-tabs/md-tabs-content-wrapper/md-tab-content[1]/div/md-content/form/div[1]/md-datepicker/div[1]/input").send_keys(Keys.BACKSPACE)
            time.sleep(1)
        except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:
            driver.refresh()
            time.sleep(1)
            click_custom_range_button()
            time.sleep(1)
            driver.find_element("xpath", "/html/body/div[2]/div[4]/md-dialog/md-tabs/md-tabs-content-wrapper/md-tab-content[1]/div/md-content/form/div[1]/md-datepicker/div").send_keys(Keys.BACKSPACE)
            time.sleep(1)
    driver.find_element("xpath","/html/body/div[2]/div[4]/md-dialog/md-tabs/md-tabs-content-wrapper/md-tab-content[1]/div/md-content/form/div[1]/md-datepicker/div[1]/input").send_keys(time_start)
    time.sleep(1)
    #Clear and Send "to" date
    for i in range(0, 11):
        driver.find_element("xpath","/html/body/div[2]/div[4]/md-dialog/md-tabs/md-tabs-content-wrapper/md-tab-content[1]/div/md-content/form/div[2]/md-datepicker/div[1]/input").send_keys(Keys.BACKSPACE)
        time.sleep(0.05)
    driver.find_element("xpath","/html/body/div[2]/div[4]/md-dialog/md-tabs/md-tabs-content-wrapper/md-tab-content[1]/div/md-content/form/div[2]/md-datepicker/div[1]/input").send_keys(time_end)
    time.sleep(1)
    #Click the "OK" button
    click_ok = True
    while click_ok == True:
        try:
            driver.find_element("xpath","/html/body/div[2]/div[4]/md-dialog/md-dialog-actions/button[2]").click()
            time.sleep(3)
            click_ok = False
        except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:
            try:
                driver.refresh()
                time.sleep(1)
                driver.find_element("xpath","/html/body/div[2]/div[4]/md-dialog/md-dialog-actions/button[2]").click()
                time.sleep(3)

                click_ok = False
            except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:
                continue
    #Click Metro
    click_sub = True
    driver.refresh()
    time.sleep(2)
    while click_sub == True:
        try:
            driver.find_element("xpath","/html/body/div[2]/div[2]/div/md-content/div/div/div[2]/trends-widget/ng-include/widget/div/div/div/div[2]/md-select").click()
            time.sleep(0.5)
            driver.find_element("xpath","/html/body/div[7]/md-select-menu/md-content/md-option[2]").click()
            time.sleep(1)
            click_sub = False
        except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:
            try:
                driver.refresh()
                time.sleep(3)
                click_and_scroll()
                time.sleep(1)
                driver.find_element("xpath","/html/body/div[2]/div[2]/div/md-content/div/div/div[2]/trends-widget/ng-include/widget/div/div/div/div[2]/md-select").click()
                time.sleep(0.5)
                driver.find_element("xpath","/html/body/div[7]/md-select-menu/md-content/md-option[2]").click()
                time.sleep(0.5)
                click_sub = False
            except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:
                continue
    #get the Geo Level
    start_date1 = datetime.strptime(time_start, "%m/%d/%Y").strftime("%Y-%m-%d")
    end_date1 = datetime.strptime(time_end, "%m/%d/%Y").strftime("%Y-%m-%d")
    encoded_string = search_keywords.replace(' ', '%20').replace('+', '%2B%20')
    time.sleep(0.5)
    
    driver.get(f"https://trends.google.com/trends/explore?date={start_date1}%20{end_date1}&geo={geo_code}&q={encoded_string}&hl=en-US")
    #download CSV file
    time.sleep(10)
    csv_file_appear = True
    while csv_file_appear == True:
        try:
            driver.find_element("xpath","/html/body/div[2]/div[2]/div/md-content/div/div/div[1]/trends-widget/ng-include/widget/div/div/div/widget-actions/div/button[1]").click()
            csv_file_appear = False
        except (NoSuchElementException, ElementClickInterceptedException, ElementNotInteractableException) as e:
            driver.refresh()
            time.sleep(30)
            continue
    time.sleep(30)
    driver.close()
```

```python
def read_and_rename_csv_file(geo_name, time_start, time_end):
    import os
    import shutil
    from datetime import datetime

    # Path to the directory containing the file
    directory = r'D:\Downloads'

    # Search for the file named "multiTimeline.csv" in the directory
    for filename in os.listdir(directory):
        if filename == 'multiTimeline.csv':

            #Clean data delete zeros and rename columns
            df = pd.read_csv(os.path.join(directory, filename), skiprows=2)
            df.Month = pd.to_datetime(df.Month).dt.strftime('%m/%d/%Y %H:%M:%S')
            df = df.set_index("Month")
            df = column_name_to_mvt_and_remove_zero_values(df, geo_name)
            df.index.name = None
            geo_date_timenow = datetime.now().strftime('%Y%m%d_%H_%M')
            new_filename = f'{geo_name}_monthly_{datetime.strptime(time_start, "%m/%d/%Y").strftime("%Y-%m-%d")}_{datetime.strptime(time_end, "%m/%d/%Y").strftime("%Y-%m-%d")}_{geo_date_timenow}.csv'
            new_path = os.path.join(r"C:\Users\tosea\GT-MVT-Monthly-City\csv_files", new_filename)
            df.to_csv(new_path)
            #Remove the file
            old_path = os.path.join(directory, filename)
            os.remove(old_path)
 
            print(f'File "{new_path}" has been saved.')

```

```python
region_code_dict = {
    "US-OH-515" : "Cincinnati OH" ,
    "US-LA-622" : "New Orleans LA"
     
}
```

```python
search_keywords = "car stolen+find stolen car+report police stolen car+insurance car stolen-dream-check"
time_start = "1/1/2015"
time_end = "12/31/2022"
```

```python
for i in range(0, 100):
    for geo_code, geo_name in region_code_dict.items():
        get_time_series_metro_GT_csv(search_keywords, time_start, time_end, geo_code)
        time.sleep(5)
        read_and_rename_csv_file(geo_name, time_start, time_end)
    time.sleep(60*60*3)
        
```
